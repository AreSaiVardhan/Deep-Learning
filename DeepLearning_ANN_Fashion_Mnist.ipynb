# Step-6: Develop ANN

- Input layer

- Hidden layer

- Output layer

- All are in sequantial

- We are passing image , image is 2D matrix form

- Means we are passing pixel values 784 pixels are passing so flatten 2D to 1D

- Every layes aftet input consider as dense layer

- But starting layer always input , last layer always output and middile layers are hidden

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense,BatchNormalization,Dropout
from tensorflow.keras.initializers import HeNormal
```

```python
model=Sequential()
################## Input #####################
# input layer   starts flatten 28x28 === 1-D
model.add(Flatten(input_shape = (28,28)))  # 28x28 = 784

################## Hidden Layer-1 ########################
# one Hidden layer
# how many neurons wants to use : 128
# which activation function wants to use : 'relu'
# bcz it avoids vanish gradients problem
model.add(Dropout((0.2)))
model.add(Dense(128,activation = 'relu',kernel_initializer = HeNormal()))

################## Hidden Layer-2 ########################
model.add(Dropout((0.2)))
model.add(Dense(128,activation = 'relu',kernel_initializer = HeNormal()))

################## Output Layer ######################
# Output layer
# how many classes are there =10
# how many neuorns : 10
# activation: softmax
model.add(Dense(10, activation = 'softmax',kernel_initializer= HeNormal()))

# Hidden layer neurons
# number of hidden layer any number
# input and output fixed
```

# Step-8 : Model Compile

```python
from re import VERBOSE
model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])

# what is the optimizer
# ADAM RMSPROP ADA GRAD
# loss : cross entropy
      #  Binary cross entropy  :  Bi classification
      #  Sparse cross entropy  :  Multi classification
# Metrics : Accuracy
```

# Step-9 : Model Fit

```python
help(model.fit)
# X : Input data
# y : Target data
# Batch size : by default 32
# epochs : 10
# verbose : It means we can able to see the answer(information)  = 1 is better
# Callback : Save model , Save the loga , Stop the training early if model will not improve much
# Validation_split : 0 to 1  - means 0 to 100%
# Validation_data :
                # data available
                # data we divided into train and test data
                # we are training model on train data
                # we are testing model on test data

                # train  ==== validation  ==== test
                # train  ==== validation (gives idea weather model is overfit or underfit)

                # test on test data (unseen data)
                # sklearn by default train test split

                # Do we have validation data available === NO
                # We have only : X_train , y_train   === again we need to divide train and validation data
                               # X_test , y_test
```

```python
X_train.shape   # 60k * 0.2 = 12k --- validation
# 48k train data
# batch size = 64
# How many batches : 48000/64 = 750  steps per epoch == if we complete 750 batchs then we complete 1 epoch
```

```python
history=model.fit(X_train,
                  y_train,
                  epochs=10,
                  batch_size=64,
                  validation_split=0.2)   # by default VERBOSE = 'auto' : auto = 1
```

```python
history.history
```

```python
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```

```python
plt.plot(history.history['loss'],label='loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```
